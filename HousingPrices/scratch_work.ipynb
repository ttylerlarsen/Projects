```python
def grapher(x_data, log_tf=False):
    if log_tf == True: 
        x = x_data
        y = full_data['SalePrice']
        #plt.xscale('log')
        plt.yscale('log')
    else:
        x = x_data / max(x_data)
        y = full_data['SalePrice'] / max(full_data['SalePrice'])
    plt.scatter(x, y)
    plt.plot([0, 1], [0, 1], color='red', linestyle='-', linewidth=2)
    plt.xlabel(x_data.name)
    plt.ylabel('Price')
    plt.show()
```

```python
def one_hot_by_col(data, col_name, prefix):
    data[col_name].fillna(0, inplace=True)
    cats = data[col_name].value_counts().index.values
    cat_type = CategoricalDtype(categories=cats)
    col_name2 = col_name+'2'
    data[col_name2] = data[col_name].astype(cat_type)
    data = pd.get_dummies(data,
                          prefix=prefix,
                          columns=[col_name2], 
                          drop_first=True)
    return data
```
